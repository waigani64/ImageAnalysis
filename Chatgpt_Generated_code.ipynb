```python
import numpy as np
import matplotlib.pyplot as plt

# Define the equilibrium constants for the reactions
K1 = 10**(-6.37)
K2 = 10**(-10.25)

# Define the total concentration of dissolved carbon dioxide
CO2_total = 1.2 * 10**(-3) # in moles per liter

# Define the pH range to simulate
pH_min = 4
pH_max = 10
pH_step = 0.1

# Initialize the arrays to store the carbonate species concentrations
H2CO3 = np.zeros((pH_max - pH_min) * 10)
HCO3 = np.zeros((pH_max - pH_min) * 10)
CO3 = np.zeros((pH_max - pH_min) * 10)

# Loop through the pH range and calculate the carbonate species concentrations
for i, pH in enumerate(np.arange(pH_min, pH_max, pH_step)):
    H = 10**(-pH)
    # Calculate the concentration of each carbonate species using the equilibrium constants
    H2CO3[i] = CO2_total / (1 + K1/H + K1*K2/H**2)
    HCO3[i] = CO2_total * K1/H / (1 + K1/H + K1*K2/H**2)
    CO3[i] = CO2_total * K1*K2/H**2 / (1 + K1/H + K1*K2/H**2)

# Plot the results
plt.plot(np.arange(pH_min, pH_max, pH_step), H2CO3, label="H2CO3")
plt.plot(np.arange(pH_min, pH_max, pH_step), HCO3, label="HCO3-")
plt.plot(np.arange(pH_min, pH_max, pH_step), CO3, label="CO32-")
plt.legend()
plt.xlabel("pH")
plt.ylabel("Concentration (M)")
plt.show()
```
