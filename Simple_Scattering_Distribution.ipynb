```python
import random
import math
import matplotlib.pyplot as plt

def scattering_simulation(num_particles, num_steps, scattering_probability):
    positions = [0] * num_particles

    for step in range(num_steps):
        for i in range(num_particles):
            if random.random() < scattering_probability:
                # Scattering occurred, change direction randomly
                angle = random.uniform(0, 2*math.pi)
                positions[i] += math.cos(angle)
            else:
                # No scattering, continue in the same direction
                positions[i] += 1

    return positions

def main():
    num_particles = 1000
    num_steps = 100
    scattering_probability = 0.2

    positions = scattering_simulation(num_particles, num_steps, scattering_probability)

    # Plotting the final positions
    plt.hist(positions, bins='auto')
    plt.xlabel('Position')
    plt.ylabel('Number of Particles')
    plt.title('Scattering Simulation')
    plt.show()

if __name__ == '__main__':
    main()
```
